```python
label_encode = LabelEncoder()
labels = label_encode.fit_transform(data["diagnosis"])
data["target"] = labels
data.drop(columns="diagnosis", axis=1, inplace=True)
```

# Data Visualization

```python
sns.countplot(x="target", data=data)
```

```python
for columns in data:
    sns.displot(x=columns, data=data)
    plt.close()
```

```python
sns.histplot(x=data.radius_mean, kde=True)
```

```python
# All of rows, first column
first_column = data.iloc[:, 0]
# All of rows, second column
second_column = data.iloc[:, 1]
```

```python
print(first_column)
print('------------')
print(second_column)
```

```python
plt.scatter(x=first_column, y=second_column)
```

```python
for columns in data:
    plt.figure()
    data.boxplot([columns])
```

```python
correlation_matrix = data.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix, cbar=True, fmt=".1f", annot=True, cmap="inferno")
plt.savefig("Breast cancer correlation heat map")
```
